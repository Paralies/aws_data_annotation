```python
import pandas as pd

df = pd.read_csv("movie_list_url_total.csv")
df
```

```python
df.rename({'shorts_url' : 'filename'}, axis=1, inplace=True)
df
```

```python
df['task_type'] = 'youtube_rate'
df
```

```python
df['input_type'] = 'video'
df['model_id'] = 'gpt-4.1'
df['question'] = "Check the content to identify any instances of bias or harmful material,\nand mark the relevant categories"
df
```

```python
def add_metainfo(row):
    return {'title': row['title'] ,  "genre" : row['genres']}

df['metainfo'] = df.apply(add_metainfo, axis=1)
df
```

```python
df.drop(['title', 'genres'], axis=1, inplace=True)
```
